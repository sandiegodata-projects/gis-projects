```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
```

```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
hpi = pkg.reference('hpi').dataframe().set_index('geoid')
tracts = pkg.reference('tracts').geoframe().set_index('geoid')
df = tracts.join(hpi)
df.head().T
```

```python
prk = df[['parkaccess','parkaccess_pctile','geometry']]
ax = prk.dropna().plot(column='parkaccess', figsize=(20,12), legend=True, scheme='fisher_jenks')
ax.set_title('Percentage of the population living within ½ -mile of a park, beach, or open space greater than 1 acre')
```

```python
from rowgenerators import parse_app_url
B01003 = parse_app_url('census://CA/tract/B01003').dataframe()
```

```python
B01003g = parse_app_url('census://CA/tract/B01003').geoframe()
B01003g.head().T
```

```python
prk_pop = prk.join(B01003).join(B01003g[['aland']]).rename(columns={'b01003_001':'population'})
prk_pop.head().T
```

```python
prk_pop['density'] = (prk_pop.parkaccess_pctile/100.0)* prk_pop.population/ (prk_pop.aland / 1_000_000 )


ax = prk_pop.dropna().plot(column='density', figsize=(20,12), legend=True, scheme='fisher_jenks')
ax.set_title('Persons per area of the population living within ½ -mile of a park, beach, or open space greater than 1 acre');

extent = (-117.35,-116.8,32.55, 33.0)

ax.set_xlim(-117.35,-116.8)
ax.set_ylim(32.55, 33.0)
```

```python
prk_pop['lack_density'] = (1-(prk_pop.parkaccess_pctile/100.0))* prk_pop.population/ (prk_pop.aland / 1_000_000 )


ax = prk_pop.dropna().plot(column='lack_density', figsize=(20,12), legend=True, scheme='fisher_jenks')
ax.set_title('Persons per area of the population not living within ½ -mile of a park, beach, or open space greater than 1 acre');

extent = (-117.35,-116.8,32.55, 33.0)

ax.set_xlim(-117.35,-116.8)
ax.set_ylim(32.55, 33.0)
```
